```python
import pandas as pd
from statsmodels.stats.multicomp import (pairwise_tukeyhsd,MultiComparison)
import matplotlib.pyplot as plt
import numpy as np
from factor_analyzer import FactorAnalyzer
```

```python
performance_file = '~/Desktop/gps.csv'
performance = pd.read_csv(performance_file)
performance.shape
```

```python
performance.columns
```

```python
df_performance = performance.loc[performance['AccelImpulse'] != 0]
```

```python
idx = df_performance.groupby(['GameID','Half','PlayerID'])['AccelImpulse'].transform(max) == df_performance['AccelImpulse']
df_max_acceleration = df_performance[idx]
df_max_acceleration.head()
```

```python

df_max_acceleration = df_max_acceleration[['GameID','Half','PlayerID','FrameID','Speed','AccelImpulse','AccelLoad']]
df_max_acceleration
index = df_max_acceleration.groupby(['GameID','Half','PlayerID'])['Speed'].transform(max) == df_max_acceleration['Speed']

df_final_performance = df_max_acceleration[index]
```

```python
df_final_performance
```

```python
game_file = '~/Desktop/games.csv'
games = pd.read_csv(game_file)
games.shape
games.head()
```

```python
games["PointsDiff"] = games["TeamPoints"] - games["TeamPointsAllowed"]
```

```python
df_games = games[['GameID','Outcome','PointsDiff']]
```

```python
dfPerformance = pd.merge(df_final_performance, 
                         df_games, left_on='GameID',right_on='GameID', how='left')
dfPerformance
```
